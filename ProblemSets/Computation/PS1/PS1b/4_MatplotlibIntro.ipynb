```python
import numpy as np
from matplotlib import pyplot as plt
```

**PROBLEM 1**

```python
def my_func(n):
    array_out = np.random.normal(size=(n,n))
    #print(array_out)
    
    aver = np.mean(array_out, axis=1)
    #print(aver)
    
    var_mu = np.var(aver)
    #print(var_mu)
    
    return var_mu
```

```python
print(my_func(2))
```

```python
def my_func2():
    result = []
    for x in range(100,1000,100):
        #print(x)
        #print(my_func(x))
        result.append(my_func(x))
      
    return result
```

```python
xlist = range(100,1000,100)
results = my_func2()
ylist = results
plt.plot(list(xlist), list(ylist))
plt.show()
```

**PROBLEM 2**

```python
xlist = np.arange(-np.pi, np.pi, 0.001)
ylist = np.sin(xlist)
ylist2 = np.cos(xlist)
ylist3 = np.arctan(xlist)
plt.plot(list(xlist), list(ylist))
plt.plot(list(xlist), list(ylist2))
plt.plot(list(xlist), list(ylist3))
plt.show()
```

**PROBLEM 3**

```python
xlist1 = np.arange(-2, 0.99, 0.01)
xlist2 = np.arange(0.99, 6, 0.01)
ylist1 = 1/(xlist1-1)
ylist2 = 1/(xlist2-1)

plt.plot(list(xlist1), list(ylist1), 'm--', lw=4)
plt.plot(list(xlist2), list(ylist2), 'm--', lw=4)
plt.ylim((-6,6))
plt.show()
```

**PROBLEM 4**

```python
x = np.linspace(0,2*np.pi,100)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
axes[0,0].plot(x, np.sin(x),'g')
axes[0,0].set_title("sin")
axes[1,0].plot(x, np.sin(2*x),'r--')
axes[1,0].set_title("sin2")
axes[0,1].plot(x, 2*np.sin(x),'b--')
axes[0,1].set_title("2sin")
axes[1,1].plot(x, 2*np.sin(2*x),'m.')
axes[1,1].set_title("2sin2")
plt.axis([0, 2*np.pi, -2, 2])
fig.suptitle("Full title")
```
