```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.special
```

**PROBLEM 2**

```python
x = np.linspace(0, 1, 200)

def my_func(n,v,x):
    return scipy.special.binom(n,v)*(x**v)*((1-x)**(n-v))

fig, ax = plt.subplots(4, 4, sharex='col', sharey='row')
for n in range(4):
    for v in range(4):
        if n >= v:
            ax[n, v].plot(x, my_func(n,v,x), label="n,v = "+str(n)+str(v))
```

**PROBLEM 3**

```python
data = np.load("MLB.npy")
print(data)
```

```python
print(type(data))
data.shape
```

```python
from numpy.polynomial.polynomial import polyfit
b, m = polyfit(data[:,1], data[:,2], 1)

plt.plot(data[:,1], b + m * data[:,1], '-')
plt.scatter(data[:,1], data[:,2], s=10, alpha=0.3)
plt.show()
```

```python
b, m = polyfit(data[:,1], data[:,0], 1)

plt.plot(data[:,1], b + m * data[:,1], '-')
plt.scatter(data[:,1], data[:,0], s=10, alpha=0.3)
plt.show()
```

**PROBLEM 5**

```python
x = np.linspace(-2, 2, 1000)
X, Y = np.meshgrid(x, x)
f = (1-X)**2 + 100*(Y - X**2)**2
```

```python
fig, ax = plt.subplots(2)
ax[0].pcolormesh(X, Y, f,  cmap="viridis")
ax[1].contour(X, Y, f, cmap="viridis")
```
